```python
import pandas as pd
```

```python
df = pd.read_csv("./results.csv", sep=',')
print(df)
```

```python
df.head()
```

```python
from pandas import DataFrame
from typing import Set, Any
def remove_others(df: DataFrame, columns: Set[Any]):
    cols_total: Set[Any] = set(df.columns)
    diff: Set[Any] = cols_total - columns
    df.drop(diff, axis=1, inplace=True)
```

```python
remove_others(df, {"Algo", "Test", "AlgInter", "Time (ns)"})
```

```python
df.head()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

# Insert

```python
dsInsert = df.loc[df['Test'] == 'Insert']
dsInsertAVL = dsInsert.loc[dsInsert['Algo'] == 'AVL']
dsInsertMap = dsInsert.loc[dsInsert['Algo'] == 'Map']
dsInsertTreap = dsInsert.loc[dsInsert['Algo'] == 'Treap']
dsInsertSplay = dsInsert.loc[dsInsert['Algo'] == 'Splay']
```

```python
First = 'AlgInter'
Second = 'Time (ns)'
plt.rcParams['figure.figsize'] = [20, 10]
plt.plot(dsInsertAVL[First], dsInsertAVL[Second] / 100 , label = 'AVL')
plt.plot(dsInsertMap[First], dsInsertMap[Second] / 100, label = 'Map')
plt.plot(dsInsertTreap[First], dsInsertTreap[Second] / 100, label = 'Treap')
plt.plot(dsInsertSplay[First], dsInsertSplay[Second] / 100, label = 'Splay')

plt.title('Time vs. Size of tree')
plt.xlabel('Size of tree (elements)')
plt.ylabel('Time (ns)')
plt.legend()
plt.savefig('Insert.png')
pass
```
