```python
import pandas as pd
```

```python
df = pd.read_csv("./results.csv", sep=',')
print(df)
```

```python
df.head()
```

```python
from pandas import DataFrame
from typing import Set, Any
def remove_others(df: DataFrame, columns: Set[Any]):
    cols_total: Set[Any] = set(df.columns)
    diff: Set[Any] = cols_total - columns
    df.drop(diff, axis=1, inplace=True)
```

```python
remove_others(df, {"Algo", "Test", "AlgInter", "Time (ns)", "Average Time Iter"})
```

```python
df.head()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

# Insert

```python
dsFind = df.loc[df['Test'] == 'Insert']
dsFindAVL = dsFind.loc[dsFind['Algo'] == 'AVL']
dsFindMap = dsFind.loc[dsFind['Algo'] == 'Map']
dsFindTreap = dsFind.loc[dsFind['Algo'] == 'Treap']
dsFindSplay = dsFind.loc[dsFind['Algo'] == 'Splay']
# dsFindAVL = dsFindAVL.sort_values(by=['AlgInter'])
```

```python
First = 'AlgInter'
Second = 'Time (ns)'
plt.rcParams['figure.figsize'] = [20, 10]
plt.plot(dsFindAVL[First], dsFindAVL[Second] / 100 , label = 'AVL')
plt.plot(dsFindMap[First], dsFindMap[Second] / 100, label = 'Map')
plt.plot(dsFindTreap[First], dsFindTreap[Second] / 100, label = 'Treap')
plt.plot(dsFindSplay[First], dsFindSplay[Second] / 100, label = 'Splay')
# plt.plot(dsFindAVL[First], np.log(dsFindAVL[First]) , label = 'LogN')
# plt.plot(dsFindAVL[First], dsFindAVL[First] , label = 'N')

# plt.plot(units_arr, test_err, label = 'train')
# plt.xlim([np.min(dsFind['Time (ns)']), np.max(dsFind['Time (ns)'])])
plt.title('Time vs. Count of iterations')
plt.xlabel('Count of iterations')
plt.ylabel('Time')
plt.legend()
plt.savefig('Insert.png')
pass
```
